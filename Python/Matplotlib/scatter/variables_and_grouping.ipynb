```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ts_data = {
    'temperature': [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2],
    'sales': [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
}
```

```python
iris_df = pd.read_csv('iris.data', names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class'])
```

```python
for label in iris_df['Class'].unique():
    print(label)
```

```python
fig = plt.figure(figsize=(8, 4))
axes = fig.subplots(1, 2)
axes[0].set_title('a. Sales over Temperature')
axes[0].set_xlabel('Temperature', fontweight='bold')
axes[0].set_ylabel('Sales', fontweight='bold')
axes[0].scatter(ts_data['temperature'], ts_data['sales'])

axes[1].set_title('b. Iris Grouping')
axes[1].set_xlabel('Sepal Length', fontweight='bold')
axes[1].set_ylabel('Sepal Width', fontweight='bold')

colors = ['#C1F7DC', '#C3D2D5', '#BDA0BC']

for i, label in enumerate(iris_df['Class'].unique()):
    plot_df = iris_df[iris_df['Class']==label]
    axes[1].scatter(plot_df['Sepal Length'], plot_df['Sepal Width'], c=colors[i], label=label)

axes[1].legend()
```
