```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris_df = pd.read_csv('iris.data', names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class'])
```

```python
feature_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

def pick_xy_labels(labels):
    """取两两特征对
    >>> feature_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
    >>> pick_xy_labels(feature_names)
    [['Sepal Length','Sepal Width']
    ['Sepal Length','Petal Length'],
    ['Sepal Length','Petal width'],
    ['Sepal width','Petal Length'],
    ['Sepal width'，'Petal width'],
    ['Petal Length''Petal Width']]
    """
    if len(labels) == 2:
        return [labels]

    r = []
    for label in labels[1:]:
        r.append([labels[0], label])

    return r + pick_xy_labels(labels[1:])

pick_xy_labels(feature_names)
```

```python
fig = plt.figure(figsize=(12, 9))
fig.subplots_adjust(wspace=0.2, hspace=0.5)
axes = fig.subplots(3, 2)

colors = ['#3A86FF', '#FB5607', '#FF006E']

for i, xy_labels in enumerate(pick_xy_labels(feature_names)):
    for j, label in enumerate(iris_df['Class'].unique()):
        plot_df = iris_df[iris_df['Class']==label]
        axes[i//2][i%2].set_xlabel(xy_labels[0])
        axes[i//2][i%2].set_ylabel(xy_labels[1])
        axes[i//2][i%2].scatter(plot_df[xy_labels[0]], plot_df[xy_labels[1]], c=colors[j], label=label)

    axes[i//2][i%2].legend()
```
