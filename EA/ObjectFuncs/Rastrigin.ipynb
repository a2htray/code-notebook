```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rastrigin(A: int):
    """经典的测试函数 Rastrigin

    函数方程式来自 https://zhuanlan.zhihu.com/p/63956652
    @:param A int 方程式参数
    @:param xValues np.ndarray 一维向量
    """
    def realFunc(xValues: np.ndarray):
        n = len(xValues)
        return A * n + np.sum(np.power(xValues, 2) - np.cos(2 * np.pi * xValues))

    return realFunc

objectFunc = rastrigin(10)
objectFunc
```

```python
# 三维空间作图，向量各分量在区间 [-5.12, 5.12]
plt.figure(figsize=(9, 9))
ax = plt.subplot(projection='3d')
x1Values = np.linspace(-5.12, 5.12, 1000)
x2Values = np.linspace(-5.12, 5.12, 1000)
x1Values, x2Values = np.meshgrid(x1Values, x2Values)
yValues = np.array([
    [
        objectFunc(np.array([x1Values[i][j], x2Values[i][j]])) for j in range(x1Values.shape[1])
    ] for i in range(x1Values.shape[0])
])

ax.plot_surface(x1Values, x2Values, yValues)
ax.set_title('Rastrigin (A=10)')
```

```python
# 二维平面下的等高线图
ctr = plt.contour(x1Values, x2Values, yValues, colors='black')
plt.contourf(x1Values, x2Values, yValues, 2)
plt.clabel(ctr)
plt.colorbar()
```
