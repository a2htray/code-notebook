```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
series = pd.Series(np.random.normal(10, 8, 20))
series
```

```python
series.plot()
```

```python
# kind
# The kind of plot to produce:
# ‘line’ : line plot (default)
# ‘bar’ : vertical bar plot
# ‘barh’ : horizontal bar plot
# ‘hist’ : histogram
# ‘box’ : boxplot
# ‘kde’ : Kernel Density Estimation plot
# ‘density’ : same as ‘kde’
# ‘area’ : area plot
# ‘pie’ : pie plot
# ‘scatter’ : scatter plot (DataFrame only)
# ‘hexbin’ : hexbin plot (DataFrame only)
series.plot(kind='bar')
```

```python
series = pd.Series(np.random.normal(2, 1, 20))
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 12))
series.plot(kind='line', ax=axes[0][0], grid=True)
series.plot(kind='bar', ax=axes[0][1], rot=360)
series.plot(kind='barh', ax=axes[0][2])
series.plot(kind='hist', ax=axes[1][0])
series.plot(kind='box', ax=axes[1][1])
series.plot(kind='kde', ax=axes[1][2])
series.plot(kind='density', ax=axes[2][0])
series.plot(kind='area', ax=axes[2][1], stacked=False)
series.plot(kind='pie', ax=axes[2][2])
fig.tight_layout()
```
