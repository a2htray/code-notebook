```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 散点图
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
ax = plt.subplot()
ax.scatter(x, y, c='r')
```

```python
data = {
    'Year': [2000, 2001, 2002, 2003],
    'Product': [10.03, 11.09, 12.17, 13.74]
}
dataframe = pd.DataFrame(data)
dataframe
```

```python
# 柱状图
## 垂直和水平
fig, [ax1, ax2] = plt.subplots(1, 2)
ax1.bar(dataframe.index, dataframe['Product'], color='red', edgecolor='yellow')
ax2.barh(dataframe.index, dataframe['Product'], color='blue', edgecolor='orange')
fig.tight_layout()
```

```python
# 等高线：在二维平面展示三维数据间的关系
x = np.linspace(-1, 1, 255)
y = np.linspace(-2, 2, 300)
z = np.sin(y[:, np.newaxis] * np.cos(x))
ax = plt.subplot()
ax.contour(x, y, z, 255)
```

```python
x = np.linspace(-1, 1, 255)
y = np.linspace(-2, 2, 300)
z = np.sin(y[:, np.newaxis] * np.cos(x))
ax = plt.subplot()
ax.contourf(x, y, z, 255)  # 会进行填充
```

```python
mu, sigma = 100, 25
fig, [ax1, ax2] = plt.subplots(2)
x = mu + sigma * np.random.randn(300)
ax1.hist(x, 20, histtype='stepfilled', facecolor='g', alpha=0.75)
ax2.hist(x, bins=[100, 150, 165, 170, 195], histtype='bar', rwidth=0.8)
```
