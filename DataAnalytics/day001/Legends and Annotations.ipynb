```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 1, 20)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
ax = plt.subplot()
ax.plot(x, y1, 'c', label='y=sin(x)')
ax.plot(x, y2, 'y', label='y=cos(x)')
ax.plot(x, y3, 'r', label='y=tan(x)')
ax.legend(loc='upper right')
```

```python
# Legend 
x = np.linspace(0, 1, 20)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
ax = plt.subplot()
plot1 = ax.plot(x, y1, 'c', label='y=sin(x)')
plot2 = ax.plot(x, y2, 'y', label='y=cos(x)')
plot3 = ax.plot(x, y3, 'r', label='y=tan(x)')
ax.add_artist(ax.legend(handles=plot1, loc='lower right'))
ax.add_artist(ax.legend(handles=plot2, loc='upper right'))
ax.add_artist(ax.legend(handles=plot3, loc='center'))
```

```python
x = np.linspace(-2.4, 0.4, 20)
y = x * x + 2 * x + 1
ax = plt.subplot()
ax.text(-1.5, 1.8, 'y=x^2+2*x+1', fontsize=14, style='italic')
ax.annotate('minima point', xy=(-1, 0), xytext=(-1, 0.3),
            horizontalalignment='center',
            verticalalignment='top',
            arrowprops=dict(arrowstyle='->'))
ax.plot(x, y)
```
