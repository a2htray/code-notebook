```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 从 0 到 3，取 6 个数值
x = np.linspace(0, 3, 6)
```

```python
fig, axes = plt.subplots(2, 2)
axes
```

```python
fig, [axes1, axes2] = plt.subplots(2, 2)
print(axes1, axes2)

axes1[0].plot(x, x, 'r')
axes1[0].set_title('Y = X')
axes1[1].plot(x, np.power(x, 3), 'b')
axes1[1].set_title('Y = X^3')
axes2[0].plot(x, np.sin(x), 'g')
axes2[0].set_title('Y = sin(X)')
axes2[1].plot(x, -2  * x, 'c')
axes2[1].set_title('Y = -2 * X')
# 保证子图间的距离
fig.tight_layout()
```

```python
# 手动绘制多个子图
plt.figure('b')
ax1 = plt.axes([0.05, 0.1, 0.4, 0.32])
ax2 = plt.axes([0.52, 0.1, 0.4, 0.32])
ax3 = plt.axes([0.05, 0.53, 0.8, 0.32])
plt.show()
```
