```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 从 0 到 3，取 6 个数值
x = np.linspace(0, 3, 6)
x
```

```python
y = np.power(x, 2)
y
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.set_title('Simple Line Plot')
ax.plot(x, y, 'r')
```

```python
# 设置 x, y Label
fig, ax = plt.subplots(figsize=(5, 3))
ax.set_title('Simple Line Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.plot(x, y, 'r')
```

```python
# set_label_position 设置 Label 位置
fig, ax = plt.subplots(figsize=(5, 3))
ax.set_title('Simple Line Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right')
ax.plot(x, y, 'r')
```

```python
# 设置线条样式、宽度
fig, ax = plt.subplots(figsize=(5, 3))
ax.set_title('Simple Line Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.xaxis.set_label_position('top')
ax.yaxis.set_label_position('right')
plot = ax.plot(x, y, 'r')[0]
plot.set_linestyle('--')
plot.set_marker('o')
plot.set_linewidth(4.0)
```
