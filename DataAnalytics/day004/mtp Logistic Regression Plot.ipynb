```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-100, 100, 2000)
xs
```

```python
ysFunc = lambda a, b: 1.0 / (1.0 + np.exp(-(a + b * xs)))
ysFunc(0, 1)
```

```python
ys = ysFunc(0, 1)
ys
```

```python
plt.plot(xs, ys)
plt.title('Logistic Regression')
plt.xlabel('x')
plt.ylabel('y')
```
