```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = torch.tensor(np.linspace(-100, 100, 10000), requires_grad=True)
```

```python
x
```

```python
y = 2 * x**2 + x + 1
```

```python
y
```

```python
plt.plot(x.data, y.data)
```

```python
# 导数
```

```python
y.backward(torch.ones_like(x))
```

```python
x.grad
```

```python
plt.plot(x.data, x.grad)
```

```python
plt.plot(x.data, y.data)
plt.plot(x.data, x.grad)
```
